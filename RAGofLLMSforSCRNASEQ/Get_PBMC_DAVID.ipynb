```python
import pandas as pd
import os

# List of URLs to download data from
urls = [
    "https://davidbioinformatics.nih.gov/data/download/chart_DAAAB5E328AF1733676049377.txt",
    "https://davidbioinformatics.nih.gov/data/download/chart_4280C34FC92A1733678959145.txt",
    "https://davidbioinformatics.nih.gov/data/download/chart_4280C34FC92A1733679065402.txt",
    "https://davidbioinformatics.nih.gov/data/download/chart_4280C34FC92A1733679159023.txt",
    "https://davidbioinformatics.nih.gov/data/download/chart_4280C34FC92A1733679280038.txt",
    "https://davidbioinformatics.nih.gov/data/download/chart_4280C34FC92A1733679390289.txt",
    "https://davidbioinformatics.nih.gov/data/download/chart_4280C34FC92A1733679548809.txt",
    "https://davidbioinformatics.nih.gov/data/download/chart_4280C34FC92A1733679682595.txt",
    "https://davidbioinformatics.nih.gov/data/download/chart_4280C34FC92A1733679778868.txt",
    "https://davidbioinformatics.nih.gov/data/download/chart_4280C34FC92A1733679872740.txt",
    "https://davidbioinformatics.nih.gov/data/download/chart_4280C34FC92A1733680018834.txt",
    "https://davidbioinformatics.nih.gov/data/download/chart_4280C34FC92A1733680115570.txt",
    "https://davidbioinformatics.nih.gov/data/download/chart_4280C34FC92A1733680221454.txt",
    "https://davidbioinformatics.nih.gov/data/download/chart_4280C34FC92A1733680302027.txt",
    "https://davidbioinformatics.nih.gov/data/download/chart_4280C34FC92A1733680363650.txt",
    "https://davidbioinformatics.nih.gov/data/download/chart_4280C34FC92A1733680439696.txt",
    "https://davidbioinformatics.nih.gov/data/download/chart_4280C34FC92A1733680536881.txt",
    "https://davidbioinformatics.nih.gov/data/download/chart_4280C34FC92A1733680698227.txt",
    "https://davidbioinformatics.nih.gov/data/download/chart_4280C34FC92A1733680797246.txt",
    "https://davidbioinformatics.nih.gov/data/download/chart_4280C34FC92A1733680894859.txt",
    "https://davidbioinformatics.nih.gov/data/download/chart_4280C34FC92A1733681052794.txt",
    # Add more URLs as needed
]

# Create the folder for saving the CSV files
folder_path = "PBMC_DAVID_Clusters"
if not os.path.exists(folder_path):
    os.makedirs(folder_path)

# Loop through the URLs and save each as a CSV with the corresponding cluster number
for idx, url in enumerate(urls, start=1):  # 'start=1' makes the cluster numbers start from 1
    # Load the data from the URL
    data = pd.read_csv(url, sep="\t", header=0)  # Ensure header is correctly interpreted
    
    # Save the entire DataFrame to a CSV first
    full_csv_path = os.path.join(folder_path, f"PBMC_DAVID_Cluster_{idx}_full.csv")
    data.to_csv(full_csv_path, index=False)
    print(f"Entire data for Cluster {idx} saved to {full_csv_path}")
    
    # Now load the CSV and select only the second column (the 'Term' column)
    term_column = data.iloc[:, 1]  # Selecting the second column directly (index 1)
    
    # Generate the CSV file path for the modified version with only the second column
    second_column_csv_path = os.path.join(folder_path, f"PBMC_DAVID_Cluster_{idx}.csv")
    
    # Save the second column to a new CSV
    term_column.to_csv(second_column_csv_path, index=False, header=False)  # header=True to include the column name
    print(f"Modified data for Cluster {idx} saved to {second_column_csv_path}")
    
    # Delete the full CSV file after saving the second column
    os.remove(full_csv_path)
    print(f"Deleted the full data file for Cluster {idx}: {full_csv_path}")
```
